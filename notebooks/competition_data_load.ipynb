```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
# FACT: This URL contains the Obermeyer et al. dataset
url = "https://gitlab.com/labsysmed/dissecting-bias/-/raw/master/data/data_new.csv"
print(f"Loading data from {url}...")
df = pd.read_csv(url)

# The dataset size is approx 48,784 rows. Loads in no time...
print(f"Data Shape: {df.shape}")

# Preprocessing
# We need to bin the risk scores into percentiles to replicate the paper's plots
df['risk_percentile'] = pd.qcut(df['risk_score_t'], 100, labels=False) + 1

df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

# This creates the histogram grouped by race
sns.histplot(
    data=df, 
    x='risk_score_t', 
    hue='race', 
    element="step",  # 'step' makes the overlap clearer than 'bars'
    stat="density",  # Normalizes so you can compare shapes even if counts differ
    common_norm=True, 
    palette={'white': 'blue', 'black': 'orange'},
    alpha=0.3
)

plt.title("Distribution of Algorithm Risk Scores by Race")
plt.xlabel("Risk Score")
plt.ylabel("Density")
plt.show()
```
